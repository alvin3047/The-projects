```python
from dateutil.parser import *
from datetime import datetime
import pandas as pd

df = open('./ICICI-Input-Case2.csv','r');
st = df.read()

st = st.replace('\n',',').split(',')

transac_type = 'Domestic'
date, trans_desc, credit,debit = None, None,None,None
li = []
counter = 3
next_name = False
curr = 'INR'
owner = None
start = 0

for j in st:
    if counter == 2:
        trans_desc = j
        counter -=1
        continue
        
    if counter == 1:
        debit = [j if j != '' else 0][0]
        counter -=1
        continue
        
    if counter == 0:
        if '-' in j:
            credit = [j if j != '' else 0][0]
            counter = 3
            if curr == -1:
                curr = trans_desc.split()[-1]
                location = trans_desc.split()[-2]
                uu = [credit.split('\n')[0] if credit.split('\n')[0] != "" else 0][0]
                li.append([str(date),trans_desc,debit,uu,curr,owner,transac_type,location])
                curr =-1
            else:
                location = trans_desc.split()[-1]
                uu = [credit.split('\n')[0] if credit.split('\n')[0] != "" else 0][0]
                li.append([str(date),trans_desc,debit,uu,curr,owner,transac_type,location])
            continue
            
        
        uu = [j if j != '' else 0][0]
        counter = 3
        if curr == -1:
            curr = trans_desc.split()[-1]
            location = trans_desc.split()[-2]
            li.append([str(date),trans_desc,debit,uu,curr,owner,transac_type,location])
            curr =-1
        else:
            location = trans_desc.split()[-1]
            li.append([str(date),trans_desc,debit,uu,curr,owner,transac_type,location])
        next_name = True
        continue
        
        
        
    if j in ['','\n',' ',None] :
        continue
    
    if "-" in j:
        next_name = False
        try: 
            date = parse(j).strftime("%d-%m-%Y")
            date = j
            counter-=1
            continue

        except:
            counter = 3
            
        
    if j in 'Domestic Transactions':
        transac_type = 'Domestic' 
        curr = 'INR'
        start = 1
        next_name = False
        continue
    if j in 'International Transactions':
        transac_type = 'International'
        curr = -1
        start =1
        next_name = False
        continue
        
    if 'Credit' in j:
        next_name = True
        start =0
        continue
    if next_name:
        next_name = False
        owner = j

    
# for i in li:
#     print(i)


df = pd.DataFrame(li)

df.columns = ['Date','Transaction Description', 'Debit','Credit','Currency','CardName','Transaction','Location']

df.sort_values(['Date'],ascending=False)

print(df)

df.to_csv('output.csv',index=False) # output
```
